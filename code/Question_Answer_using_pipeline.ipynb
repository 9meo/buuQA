```python
from transformers import pipeline

oracle = pipeline(model="deepset/roberta-base-squad2")
oracle(question="Where do I live?", context="My name is Wolfgang and I live in Berlin")
```

```python
context='มนุษย์เงินเดือนและผู้มีรายได้ที่ต้องเสียภาษีเงินได้บุคคลธรรมดาในปี 2567 คงถูกอกถูกใจกับโครงการ Easy E-Receipt ซึ่งให้เรานำค่าใช้จ่ายจากการซื้อสินค้าและบริการในช่วงระหว่างวันที่ 1 มกราคม - 15 กุมภาพันธ์ 2567 มาใช้ลดหย่อนภาษีได้สูงสุดถึง 50,000 บาท แต่จะได้รับเงินคืนภาษีเท่าไรนั้นขึ้นอยู่กับฐานภาษี ดังนี้'
question='ลดหย่อนภาษีได้สูงสุดกี่บาท'
```

```python
oracle(question=question, context=context)
```

```python
thaiqa(question=question, context=context)
```
